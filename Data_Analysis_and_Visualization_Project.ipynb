```python
numerical_columns = ['age', 'time_spent', 'income']

mean_values = df[numerical_columns].mean()
mode_values = df[numerical_columns].mode().iloc[0]  # Mode might have multiple values, so we take the first one
median_values = df[numerical_columns].median()

ct_statistics_df = pd.DataFrame({
    'Mean': mean_values,
    'Mode': mode_values,
    'Median': median_values
}, index=numerical_columns)

ct_statistics_df
```

#Measures of Dispersion


```python
range_values = df[numerical_columns].max() - df[numerical_columns].min()
variance_values = df[numerical_columns].var()
std_dev_values = df[numerical_columns].std()

disp_statistics_df = pd.DataFrame({
    'Range': range_values,
    'Variance': variance_values,
    'Standard Deviation': std_dev_values
}, index=numerical_columns)

disp_statistics_df
```

##Mean Absolute Deviation

```python
# Mean Deviation of Ages
age_mean = df['age'].mean()
sum_age_diff = 0
for age in df['age']:
  sum_age_diff += abs(age_mean - age)
mean_deviation_age = sum_age_diff / df['age'].count()
print('Mean Deviation of Ages =', mean_deviation_age)

# Mean Deviation of Incomes
income_mean = df['income'].mean()
sum_income_diff = 0
for income in df['income']:
  sum_income_diff += abs(income_mean - income)
mean_deviation_income = sum_income_diff / df['income'].count()
print('Mean Deviation of Incomes =', mean_deviation_income)

# Mean Deviation of Time Spent
time_spent_mean = df['time_spent'].mean()
sum_time_spent_diff = 0
for time_spent in df['time_spent']:
  sum_time_spent_diff += abs(time_spent_mean - time_spent)
mean_deviation_time_spent = sum_time_spent_diff / df['time_spent'].count()
print('Mean Deviation of Time Spent =', mean_deviation_time_spent)
```

**Research Question 2:** Is there a significant difference in income between genders?
This question looks at the difference in a numerical variable (income) across two groups defined by a categorical variable (gender). The appropriate test for this is the Independent T-Test.

**Null Hypothesis (H0): **There is no significant difference in income between genders.

**Alternative Hypothesis (H1):** There is a significant difference in income between genders.


**Conducting the Independent T-Test:**

```python
from scipy.stats import ttest_ind

# Assuming 'df' is your DataFrame and 'income' and 'gender' are column names
male_income = df[df['gender'] == 'Male']['income']
female_income = df[df['gender'] == 'Female']['income']

t_stat, p_value = ttest_ind(male_income, female_income, equal_var=False)  # Using Welch's t-test due to possible unequal variances

# Print the results
print("Independent Samples T-Test Results:\n")
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis. There is a significant difference in income between genders.")
else:
    print("\nFail to reject the null hypothesis. There is no significant difference in income between genders.")
```
